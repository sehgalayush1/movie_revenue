# Test Data Feature Extraction

### Adding actors and directors data to test data

```python
import numpy as np
import pandas as pd
```

```python
test_file = 'data/test_data.csv'
test_data = pd.read_csv(test_file)
```

```python
actors_csv = 'data/BollywoodActorRanking.csv'
directors_csv = 'data/BollywoodDirectorRanking.csv'
actors_df = pd.read_csv(actors_csv)
directors_df = pd.read_csv(directors_csv)
```

```python
test_data
```

```python
test_data['actors'] = test_data['actors'].str.split(',')
test_data['actors'].values
```

```python
test_data['actors'].values
```

```python
actor_data = list()
for actors in test_data['actors']:
    actors = [s.strip() for s in actors]
    new_df = actors_df.loc[actors_df['actorName'].isin(actors)]
    actor_data.append([new_df['movieCount'].mean(), new_df['ratingSum'].mean(), new_df['normalizedMovieRank'].mean(), new_df['googleHits'].mean(), new_df['normalizedGoogleRank'].mean(), new_df['normalizedRating'].mean()])
print(actor_data)
```

```python
actor_data = pd.DataFrame(data=actor_data, columns=['actors_movieCount', 'actors_ratingSum', 'actors_normalizedMovieRank',
                'actors_googleHits', 'actors_normalizedGoogleRank', 'actors_normalizedRating'])
```

```python
actor_data = actor_data.fillna(actor_data.mean())
actor_data
```

```python
test_data['directors'] = test_data['directors'].str.split(', ')
test_data['directors']
```

```python
director_data = list()
for directors in test_data['directors']:
    directors = [s.strip() for s in directors]
    new_df = directors_df.loc[directors_df['directorName'].isin(directors)]
    director_data.append([new_df['movieCount'].mean(), new_df['ratingSum'].mean(), new_df['normalizedMovieRank'].mean(), new_df['googleHits'].mean(), new_df['normalizedGoogleRank'].mean(), new_df['normalizedRating'].mean()])
print(director_data)
```

```python
director_data = pd.DataFrame(data=director_data, columns=['movieCount', 'ratingSum', 'normalizedMovieRank',
                'googleHits', 'normalizedGoogleRank', 'normalizedRating'])
director_data
```

```python
new_list = list()
for l in directors_df.iloc[0:9, 2:].values:
    new_list.append(l)
new_list
```

```python
len(new_list)
```

```python
director_data.loc[(director_data.isnull().all(axis=1))] = new_list
director_data
```

```python
director_data.columns = ['directors_' + str(col) for col in director_data.columns]
director_data
```

```python
test_data = test_data.join(actor_data)
test_data = test_data.join(director_data)
test_data
```

```python
test_data.to_csv('data/final_test_data.csv')
```
